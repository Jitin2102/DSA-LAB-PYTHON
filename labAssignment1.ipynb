### **Ques 1: Write a Python program to compute the monthly salary of an employee based on the following inputs:** 
**Basic Salary (float)**   
**HRA (Housing Rent Allowance) = 20% of Basic Salary**  
**DA (Dearness Allowance) = 50% of Basic Salary** 
**Tax = 10% of Gross Salary**  
**Net Salary = Gross Salary - Tax**   
**Display: Gross Salary, Tax Amount and Net Salary**   
**Use appropriate type casting and operators in your calculations.**  


```python
basicSalary = float(input("Enter the basic salary: "))
hra = 0.20 * basicSalary
da = 0.45 * basicSalary
grossSalary = basicSalary + hra + da
print("Gross Salary is:", grossSalary)
taxAmount = 0.30 * grossSalary
netSalary = grossSalary - taxAmount
print("Tax Amount is:", taxAmount)
print("Net Salary is:", netSalary)
```

### **Ques 2 : Write a Python program to input a student's marks and determine their grade based on the following conditions:**
**Marks >= 90: Grade A**  
**Marks >= 75 and < 90: Grade B**  
**Marks >= 50 and < 75: Grade C**  
**Marks < 50: Fail**


```python
marks = float(input("Enter the marks:"))
if marks >= 90:
    grade = 'A'
elif marks >= 75 and marks < 90:
    grade = 'B'
elif marks >= 50 and marks < 75:
    grade = 'C'
else:
    grade = 'F'
print("The grade is: ", grade)
```

### **Ques 3 :Write a Python program to check if a given year is a leap year.**  
**Conditions:**  
**(a)	A year is a leap year if it is divisible by 4.**  
**(b)	However, years divisible by 100 are not leap years unless they are also divisible by 400.**


```python
year = int(input("Enter a year: "))
if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(year, "is a leap year.")
else:
    print(year, "is not a leap year.")
```

### **Ques 4 :Write a Python program to calculate the sum of all odd numbers up to a given number n.**

```python
n = int(input("Enter a number: "))
sum = 0
for i in range(1, n + 1,2):
    sum+=i
print("The sum of all odd numbers from 1 to", n, "is:", sum)
```

### **Ques 5 : Write a Python program to reverse a given integer using a while loop.**


```python
n = int(input("Enter a number: "))
rev = 0
temp = n
while(temp):
    rev = rev * 10 + temp % 10
    temp = temp // 10
print("The reverse of", n, "is:", rev)
```

### **Ques 6 : Write a Python program to find the first prime number greater than a given number n.**

```python
no = int(input("Enter a number: "))

def isFirstPrime(num):
    if num <= 1:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True
n = no
while n > 0:
    if isFirstPrime(n + 1):
        print("The first prime no. greater than",no,"is:", n + 1)
        break
    n += 1
```

### **Ques 7: Write a Python program that prints all numbers from 1 to n, skipping those that are multiples of 3.**

```python
n = int(input("Enter a number: "))
for i in range(1,n):
    if(i % 3)==0:
        continue
    print(i,end=' ')
```

### **Ques 9 : Write a program to count number of vowels in a string.**

```python
s = input("Enter a string: ")
cnt =0
for char in s:
    if char.lower() in 'aeiou':
        cnt += 1
print("Vowels in entered string ",s,"are ",cnt,".")
```
